# Support Vector Machine Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC, SVR
```

```python
from sklearn.datasets import load_iris, fetch_california_housing
```

```python
iris = load_iris()
```

```python
x_classification =pd.DataFrame(iris.data,columns=iris.feature_names)
print(x_classification)
```

```python
y_classification = iris.target
print(y_classification)
```

```python
california = fetch_california_housing()
```

```python
x_regression =pd.DataFrame(california.data,columns =california.feature_names )
print(x_regression)
```

```python
y_regression =california.target
print(y_regression)
```

```python
# Regression
xr_train, xr_test, yr_train, yr_test=train_test_split(x_regression, y_regression, test_size=0.2, random_state=42) 
```

```python
xc_train,xc_test, yc_train, yc_test=train_test_split(x_classification,y_classification, test_size=0.2, random_state=42)
```

```python
scaler=StandardScaler()  
```

```python
xr_train=scaler.fit_transform(xr_train) 
xr_test=scaler.transform(xr_test) 
```

```python
xc_train= scaler.fit_transform(xc_train) 
xc_test= scaler.transform(xc_test) 
```

```python
svc_model=SVC(kernel="linear")
svc_model.fit(xc_train,yc_train)
```

```python
svr_model =SVR(kernel="linear")
svr_model.fit(xc_train,yc_train)
```

```python
yc_pre = svc_model.predict(xc_test)
```

```python
print(yc_pre)
```

```python
yr_pre = svr_model.predict(xc_test)
print(yr_pre)
```
